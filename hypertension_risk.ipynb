## **一、背景描述**

本数据集旨在探索高血压风险的预测因素，通过人口统计学和健康相关属性的数据支持**高血压风险**的评估和预防策略的开发。

数据集包含**13**个特征，涵盖性别、年龄、吸烟习惯、是否服用降压药、是否患有糖尿病、胆固醇水平、收缩压和舒张压、体重指数（BMI）、心率、血糖水平以及高血压风险标签（0 表示低风险，1 表示高风险）。这些数据为开发预测模型提供了全面的基础。

数据来源：https://www.heywhale.com/mw/dataset/6736f8d20bb59b78302c1397

## **二、数据说明**

- male	性别，0 表示女性，1 表示男性
- age	个体的年龄
- currentSmoker	当前吸烟状态，0 表示非吸烟者
- cigsPerDay	每天吸烟的支数（整数）
- BPMeds	是否服用降压药，0 表示未服用
- diabetes	糖尿病状态，0 表示无糖尿病
- totChol	总胆固醇水平（mg/dL）
- sysBP	收缩压（mmHg）
- diaBP	舒张压（mmHg）
- BMI	体重指数（体重/身高²，kg/m²）
- heartRate	心率（每分钟心跳数）
- glucose	血糖水平（mg/dL）
- Risk	高血压风险标签，**0 表示低风险，1 表示高风险**

### **1. 吸烟**

吸烟对高血压的影响包括直接和间接两方面。直接影响是尼古丁刺激交感神经系统，导致心率加快和血管收缩，从而升高血压；间接影响包括引发氧化应激、慢性炎症、血脂异常和胰岛素抵抗，进一步损害血管健康并加重高血压。长期吸烟会显著增加心血管疾病的风险，

### **2. 胆固醇指标**


| 指标               | 理想水平 | 临界高风险 | 高风险 |
|:------------------:|:--------:|:----------:|:------:|
| 总胆固醇（Total）  | < 200    | 200-239    | ≥ 240  |


### **3. 收缩压指标**

#### 脉压差：收缩压与舒张压的差值称为脉压差，正常范围为30-50 mmHg。
- 脉压差增大（> 50 mmHg）可能提示动脉硬化。
- 脉压差减小（< 30 mmHg）可能提示心脏泵血功能下降或血容量不足。

####血压分类：
- 正常血压：sysBP < 120 mmHg 且 diaBP < 80 mmHg
- 高血压前期：sysBP 120-139 mmHg 或 diaBP 80-89 mmHg
- 高血压：sysBP ≥ 140 mmHg 或 diaBP ≥ 90 mmHg

### **4. 体重指数指标**

定义：BMI是通过体重（公斤）除以身高（米）的平方计算得出的数值。

公式：BMI = 体重（kg）/身高（m）**2



| BMI范围      | 分类     | 健康风险                                   |
|:------------:|:--------:|:-----------------------------------------:|
| < 18.5       | 体重过轻 | 营养不良、免疫力下降                       |
| 18.5 - 24.9  | 正常体重 | 健康风险最低                               |
| 25.0 - 29.9  | 超重     | 增加高血压、糖尿病等风险                   |
| ≥ 30.0       | 肥胖     | 显著增加心血管疾病、糖尿病等风险           |

### **5. 心率指标**

心率因年龄、性别、身体状况和活动水平而异。以下是不同人群的静息心率正常范围：


| 人群	         | 正常静息心率（bpm）|
|:-------------:|:---------------:|
| 成年人	       | 60 - 100        |
| 运动员或健康人群 | 40 - 60         |
| 儿童（6-15岁）  | 70 - 100        |
| 老年人         | 60 - 100         |

- 较低的静息心率通常与较好的心血管健康相关。
- 静息心率过高可能提示潜在健康问题，如高血压、心脏病或代谢综合征。

### **6. 血糖水平指标**

| 测量方法             | 正常范围                     | 糖尿病前期               | 糖尿病                   |
|----------------------|------------------------------|--------------------------|--------------------------|
| 空腹血糖             | 70 - 99 mg/dL (3.9 - 5.5 mmol/L) | 100 - 125 mg/dL (5.6 - 6.9 mmol/L) | ≥ 126 mg/dL (≥ 7.0 mmol/L) |
| 餐后2小时血糖        | < 140 mg/dL (< 7.8 mmol/L)   | 140 - 199 mg/dL (7.8 - 11.0 mmol/L) | ≥ 200 mg/dL (≥ 11.1 mmol/L) |
| 糖化血红蛋白（HbA1c） | < 5.7%                      | 5.7% - 6.4%              | ≥ 6.5%                  |

#### 糖尿病与高血压的流行病学
- 共同高发：糖尿病患者中高血压的发病率显著高于非糖尿病人群。据统计，约60%-80%的2型糖尿病患者同时患有高血压。

- 相互影响：糖尿病和高血压互为危险因素，共同增加心脑血管疾病的风险。

### **综上所述**
数据集中的特征（如年龄、性别、吸烟习惯、胆固醇水平、BMI、血糖水平等）均与高血压风险密切相关。接下来，我们将首先对数据进行清洗，以确保数据质量。随后，通过探索性数据分析（EDA）和可视化手段，逐步发掘各个因子对高血压风险的影响，并利用机器学习模型评估其预测能力。最终，结合分析结果，为高血压的预防和干预提供科学依据和个性化建议。

#### 3.3.1 胆固醇

```python
#| 指标               | 理想水平 | 临界高风险 | 高风险 |
#| 总胆固醇（Total）  | < 200    | 200-239    | ≥ 240  |
# 低风险L， 临界高风险M， 高风险H
df_data['胆固醇风险'] = df_data['胆固醇水平（mg/dL）'].apply(lambda x: '低风险' if(x<200) else('中风险' if(x>=200 and x<=239) else '高风险'))
df_data[['胆固醇水平（mg/dL）','胆固醇风险']].head()
```

#### 3.3.2 脉压差，血压


```python
# 脉压差 = 收缩压 - 舒张压
# 脉压差增大（> 50 mmHg）可能提示动脉硬化。 脉压差减小（< 30 mmHg）可能提示心脏泵血功能下降或血容量不足。
df_data['脉压差'] = (df_data['收缩压（mmHg）'] - df_data['舒张压（mmHg）']).apply(lambda x: '偏大' if(x>50) else('偏小' if(x<30) else '正常'))
df_data[['收缩压（mmHg）','舒张压（mmHg）','脉压差']].head(5)
```

```python
# 血压
######### FUNC1 ###########
# 根据舒张压和收缩压范围判断高血压
def HT_class(df_data):
    # 获取收缩压和舒张压
    systolic = df_data['收缩压（mmHg）']
    diastolic = df_data['舒张压（mmHg）']

    # 判断低血压
    if systolic < 90 or diastolic < 60:
        return '低血压'
    # 判断正常血压
    elif systolic < 120 and diastolic < 80:
        return '正常血压'
    # 判断高血压前期
    elif (120 <= systolic < 140) or (80 <= diastolic < 90):
        return '高血压前期'
    # 判断高血压
    elif systolic >= 140 or diastolic >= 90:
        return '高血压'
    # 异常情况
    else:
        raise ValueError('请检查输入的收缩压或舒张压')
############################


df_data['血压'] = df_data.apply(HT_class, axis=1)
df_data[['收缩压（mmHg）','舒张压（mmHg）','血压']].head(5)
```

```python
def standardize(data):
    """
    对一维数据进行标准化（减去均值再除以标准差）

    参数:
    data (list/np.ndarray/pd.Series): 输入的一维数据

    返回:
    np.ndarray: 标准化后的一维数据
    """
    data = np.array(data)  # 将输入数据转换为NumPy数组，以便进行统一处理
    mean = np.mean(data)  # 计算均值
    std = np.std(data)  # 计算标准差
    if std == 0:  # 如果标准差为0，直接返回原数据（避免除以0的错误）
        return data
    standardized_data = (data - mean) / std  # 进行标准化计算
    return standardized_data
```

```python
# 仅对数值型变量进行标准化处理，因为分类型数据里面仅有0和1两个值
df_data_model['age'] = standardize(df_data_model['age'])
df_data_model['BMI'] = standardize(df_data_model['BMI'])
df_data_model['totChol'] = standardize(df_data_model['totChol'])
df_data_model['heartRate'] = standardize(df_data_model['heartRate'])
```

#### 模型评估函数

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, roc_curve, auc


def calculate_metrics(y_test, y_pred, y_pred_proba=None):
    """
    计算并展示分类模型的各项评估指标

    Parameters:
    -----------
    y_test : array-like
        真实标签
    y_pred : array-like
        预测标签
    y_pred_proba : array-like, optional
        预测为正类的概率，用于绘制ROC曲线
    """
    # 1. 计算基础指标
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # 打印指标
    print("模型评估指标:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")

    # 计算混淆矩阵
    conf_matrix = confusion_matrix(y_test, y_pred)
    sns.heatmap(conf_matrix, annot=True, annot_kws={'size': 12},fmt='d', cmap='Blues')

    # 设置 xtick 和 ytick
    plt.xticks([0.5, 1.5], ['Low Risk', 'High Risk'], fontsize=10)
    plt.yticks([0.5, 1.5], ['Low Risk', 'High Risk'], fontsize=10, rotation=90)

    plt.xlabel('Predicted', fontsize=12)
    plt.ylabel('True', fontsize=12)
    plt.show()
    # 3. 如果提供了概率预测值，绘制ROC曲线
    if y_pred_proba is not None:
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(4, 4))
        plt.plot(fpr, tpr, color='darkorange', lw=2,
                 label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) Curve')
        plt.legend(loc="lower right")
        plt.show()

        print(f"ROC AUC: {roc_auc:.4f}")

    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'confusion_matrix': conf_matrix,
        'roc_auc': roc_auc if y_pred_proba is not None else None
    }
```

#### 逻辑回归

```python
# -- 找最佳参数
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# 定义参数网格
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga'],
    'max_iter':[100, 500, 1000]
}

# 创建模型
model = LogisticRegression(max_iter=1000)

# 网格搜索
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# 输出最优参数
print("Best parameters:", grid_search.best_params_)
```

```python
# -- 模型评估
best_model = LogisticRegression(
    C=grid_search.best_params_['C'],
    penalty=grid_search.best_params_['penalty'],
    solver=grid_search.best_params_['solver'],
    max_iter=grid_search.best_params_['max_iter']
)

# 训练模型
best_model.fit(X_train, y_train)

# 对于使用sklearn的模型
y_pred = best_model.predict(X_test)
y_pred_proba = best_model.predict_proba(X_test)[:, 1]  # 获取正类的概率

# 计算所有指标
metrics = calculate_metrics(y_test, y_pred, y_pred_proba)

```
